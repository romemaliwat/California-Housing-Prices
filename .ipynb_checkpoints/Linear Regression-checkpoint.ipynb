```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('housing.csv')
data
```

```python
data.info()
```

```python
data.dropna(inplace = True)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(['median_house_value'], axis = 1) # X is dataframe without the target variable (in this case, the 'median house value' column)
y = data['median_house_value'] # y is the target variable (in this case, 'median house value')
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
train_data = X_train.join(y_train)
train_data
```

```python
train_data.hist(figsize = (15, 8))
```

```python
train_data.corr(numeric_only = True)
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = 'YlGnBu')
```

```python
train_data['total_rooms'] = np.log(train_data['total_rooms'] - 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] - 1)
train_data['population'] = np.log(train_data['population'] - 1)
train_data['households'] = np.log(train_data['households'] - 1)
```

```python
train_data.hist(figsize = (15, 8))
```

```python
train_data.ocean_proximity.value_counts()
```

```python
pd.get_dummies(train_data['ocean_proximity'], dtype = int)
```

```python
train_data = train_data.join(pd.get_dummies(train_data['ocean_proximity'], dtype = int)).drop(['ocean_proximity'], axis = 1)
```

```python
train_data
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = 'YlGnBu')
```

```python
plt.figure(figsize = (15,8))
sns.scatterplot(x = 'latitude', y = 'longitude', data = train_data, hue = 'median_house_value', palette = 'coolwarm')
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
train_data['household_rooms'] = train_data['total_rooms'] / train_data['households']
```

```python
plt.figure(figsize = (15, 8))
sns.heatmap(train_data.corr(numeric_only = True), annot = True, cmap = 'YlGnBu')
```

```python
from sklearn.linear_model import LinearRegression
X_train, y_train = train_data.drop(['median_house_value'], axis = 1), train_data['median_house_value']
reg = LinearRegression()
reg.fit(X_train, y_train)
```
